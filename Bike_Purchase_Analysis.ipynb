## Exploratory Data Analysis (EDA)
Different visualizations are used to analyze customer behavior and identify patterns in bike purchases.

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\singh\OneDrive\Desktop\project\bike_purchase.csv")
df.head()
```

### Insight:
- Majority of customers did not purchase bikes.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Purchased Bike', data=df)
plt.show()
```

### Insight:
- Male customers show slightly higher bike purchase rates compared to females.

```python
sns.countplot(x='Gender', hue='Purchased Bike', data=df)
plt.show()
```

### Insight:
- Customers with higher income are more likely to purchase bikes.

```python
sns.boxplot(x='Purchased Bike', y='Income', data=df)
plt.show()
```

### Insight:
- Middle-aged individuals are more likely to buy bikes.

```python
sns.histplot(df['Age'], bins=20)
plt.show()
```

## Conclusion
The analysis shows that income, age, and demographic factors influence bike purchase decisions. Middle-aged individuals and higher income groups are more likely to purchase bikes. The insights can help businesses target potential customers more effectively.

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

### Insight:
- Some variables like income and age show moderate influence on bike purchase.
- Most variables have low correlation, indicating independent factors.
